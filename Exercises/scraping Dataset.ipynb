```
#setting the path to a variable

elon_path = "/content/downloads/elonmusk"
jeff_path = "/content/downloads/jeffbezos"
```

```
#To check out if the images are present
import os
os.listdir(elon_path)[:5] #what this does is, it lists out the first 5 files present in the messi folder
os.listdir(jeff_path)[:6]
```

```
# function to remove images that aren't jpgs

def clean_dir(PATH):
    for file in os.listdir(PATH):
        if file.endswith('.jpg'):
            pass
        else:
            os.remove(f'{PATH}/{file}')#if the file isnt a jpg file, it is removed/deleted from the directory

# function to trim the folder down to 100 images

def shape_dir(PATH):
    count = 0
    for file in os.listdir(PATH):
        count += 1
        if count > 100: #if the count of the files in the directory exceeds 100, then the surplus files are deleted 
            os.remove(f'{PATH}/{file}')
    print('finished')
```

```
clean_dir(elon_path)
clean_dir(jeff_path)

shape_dir(elon_path)
shape_dir(jeff_path)
```

```
os.getcwd()

new_name = ['elon', 'jeff']

def parse_images(path, new_name):
    for i, filename in enumerate(os.listdir(path)):

        #to rename the file...use os.rename() 
        #syntax is:= os.rename(source address, destination address)

        os.rename(path + "/" + filename, path + "/" + new_name +  str(i) + ".jpg") 

parse_images(elon_path, new_name[0])
parse_images(jeff_path, new_name[1])
```

```
source = ['/content/downloads/elonmusk', '/content/downloads/jeffbezos']
train_dest = ['/content/data/train/elonmusk', '/content/data/train/jeffbezos']
valid_dest = ['/content/data/valid/elonmusk', '/content/data/valid/jeffbezos']
```

```
#now we need to move the files in the messi and ronaldo subfolders of the download folder to
#the the train and test folder of the messi and ronaldo.
#shutil.move() provides the soln,i.e., for moving files

#syntax: shutil.move(source, destination)

import shutil

# Creating our function to split into Train, Valid sets
#Splitting the files is done in the ratio 80:20. i.e., 80 for training and 20 for testing

def split_images(source, train_dest, valid_dest, f_name):
    files = os.listdir(source)
    iter_t = 80
    iter_v =20
    for i in range(iter_t):
        shutil.move(source + '/' + f_name + str(i) + '.jpg', train_dest)
    for i in range(iter_v):
        shutil.move(source + '/' + f_name +  str(i + 80) + '.jpg', valid_dest)
```

```
# Calling the functions

# elonmusk
split_images(source[0], train_dest[0], valid_dest[0], f_name='elonmusk')

# jeffbezos
split_images(source[1], train_dest[1], valid_dest[1], f_name='jeffbezos')
```
