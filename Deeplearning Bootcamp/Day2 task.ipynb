```
#setting the path to a variable
messi_path = "/content/downloads/messi"
ronaldo_path = "/content/downloads/ronaldo"
```

```
#To check out if the images are present
import os
os.listdir(messi_path)[:5] #what this does is, it lists out the first 5 files present in the messi folder
os.listdir(ronaldo_path)[:6]
```

```
# function to remove images that aren't jpgs
def clean_dir(PATH):
    for file in os.listdir(PATH):
        if file.endswith('.jpg'):
            pass
        else:
            os.remove(f'{PATH}/{file}')#if the file isnt a jpg file, it is removed/deleted from the directory

# function to trim the folder down to 100 images
def shape_dir(PATH):
    count = 0
    for file in os.listdir(PATH):
        count += 1
        if count > 100: #if the count of the files in the directory exceeds 100, then the surplus files are deleted 
            os.remove(f'{PATH}/{file}')
    print('finished')
```

```
clean_dir(messi_path)
clean_dir(ronaldo_path)

shape_dir(messi_path)
shape_dir(ronaldo_path)
```

```
os.getcwd()

new_name = ['messi', 'ronaldo']

def parse_images(path, new_name):
    for i, filename in enumerate(os.listdir(path)):
        #to rename the file...use os.rename() 
        #syntax is:= os.rename(source address, destination address)
        os.rename(path + "/" + filename, path + "/" + new_name +  str(i) + ".jpg") 

parse_images(messi_path, new_name[0])
parse_images(ronaldo_path, new_name[1])
```

Now we need to split the data into Test and Train folders.Thus, we need to create folders of similar structure<br><br>4.jpg

```
#mkdir>>make 

%mkdir data
%mkdir data/train | mkdir data/valid
%mkdir data/train/messi | mkdir data/train/ronaldo
%mkdir data/valid/messi | mkdir data/valid/ronaldo
```

```
source = ['/content/downloads/messi', '/content/downloads/ronaldo']
train_dest = ['/content/data/train/messi', '/content/data/train/ronaldo']
valid_dest = ['/content/data/valid/messi', '/content/data/valid/ronaldo']
```

```
#now we need to move the files in the messi and ronaldo subfolders of the download folder to
#the the train and test folder of the messi and ronaldo.
#shutil.move() provides the soln,i.e., for moving files

#syntax: shutil.move(source, destination)
import shutil

# Creating our function to split into Train, Valid sets
#Splitting the files is done in the ratio 80:20. i.e., 80 for training and 20 for testing

def split_images(source, train_dest, valid_dest, f_name):
    files = os.listdir(source)
    iter_t = 80
    iter_v =20
    for i in range(iter_t):
        shutil.move(source + '/' + f_name + str(i) + '.jpg', train_dest)
    for i in range(iter_v):
        shutil.move(source + '/' + f_name +  str(i + 80) + '.jpg', valid_dest)
```

```
# Calling the functions

# Messi
split_images(source[0], train_dest[0], valid_dest[0], f_name='messi')

# Ronaldo
split_images(source[1], train_dest[1], valid_dest[1], f_name='ronaldo')
```

```
base_dir = "data"
train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'valid')

train_messi_dir = os.path.join(train_dir, 'messi')
train_ronaldo_dir = os.path.join(train_dir, 'ronaldo')

valid_messi_dir = os.path.join(validation_dir, 'messi')
valid_ronaldo_dir = os.path.join(validation_dir, 'ronaldo')
```

```
print("Training and Validation Images of Messi are {} and {} respectively".format(
    len(os.listdir(train_messi_dir)),
    len(os.listdir(valid_messi_dir))
))
```

```
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=64,
    class_mode='binary'
  )

validation_generator = test_datagen.flow_from_directory(
    validation_dir,
    target_size=(150, 150),
    batch_size=64,
    class_mode='binary'
  )
```

```
import matplotlib.pyplot as plt

index_to_class = {0:"messi",1:"ronaldo"}
plt.figure(figsize=(10, 10))
for imagebatch, labelbatch in train_generator:
  print("Image Batch Shape: " + str(imagebatch.shape))
  print("Label Batch Shape: " + str(labelbatch.shape))
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(imagebatch[i])
    plt.title("Label: " + index_to_class[labelbatch[i]])
    plt.axis("off")
  break
```
